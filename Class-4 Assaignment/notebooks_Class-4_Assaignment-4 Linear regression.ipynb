```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Height-Weight Data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
x = df[['Height']] # independent feature
y = df['Weight']   # dependent feature
```

```python
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x,y, marker='*', color='yellowgreen')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# split data
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3)
```

```python
lr = LinearRegression() # linear regression model object
```

```python
lr.fit(xtrain, ytrain) # fitting model
```

```python
df['Predict Weight'] = lr.predict(x)
```

```python
df
```

## go for the comparison

```python
plt.xlabel('Height')
plt.ylabel('Weight')

plt.scatter(x,y,marker='*',color='yellowgreen', label='raw')

plt.scatter(x,df['Predict Weight'], color='green', label='Pred')

plt.legend()
plt.show()
```

```python
pred = lr.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mse = mean_squared_error(ytest,pred)
```

```python
mse
```

```python
mae = mean_absolute_error(ytest,pred)
```
